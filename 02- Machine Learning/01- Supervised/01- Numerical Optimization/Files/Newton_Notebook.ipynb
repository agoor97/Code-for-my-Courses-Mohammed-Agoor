```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import fsolve   ## to solve the equaion
```

#### Optimization_1st Example

```python
def f(x):  ## Define the Function of the equation
    return 3*x**2 + 4*x - 1

x_new = np.linspace(-5, 5)  ## random values for function
y = f(x=x_new)  ## call the function and get the y of these values

## Solution of the Equation, Can be found analytically of numerically
sol = [0.21525044, -1.54858377]   ## this is the solution, get it analytically


## Plot the function
plt.figure(figsize=(10, 6))
plt.plot(x_new, y, label='Function')
plt.axhline(xmin=-4, xmax=4, y=0, c='r', label='x-axis')
plt.scatter(sol[0], f(sol[0]), marker='o', color='blue', s=100, label='Solution')
plt.scatter(sol[1], f(sol[1]), marker='o', color='blue', s=100)
plt.xlabel('x values')
plt.ylabel('y values')
plt.legend()
plt.show()
```

#### Optimization_2nd Example

```python
## what about --> ^+  = _ (+)
def f2(x):
    return 3*x**2 + 4*x - 1 - 4*np.log10(x+3)

## Define some random range of x values
x_new2 = np.linspace(-2, 2)
y2 = f2(x_new2)  ## Get the y by calling the function


## solution of this function can't be finded analytically 
## the best way is to find it numerically
val_1 = fsolve(f2, -2)  ## intital value to start optimizing  --> -2
val_2 = fsolve(f2, 2)   ## intital value to start optimizing  --> 2

print('Value 1', val_1)
print('Value 2', val_2)
print('**'*20)

## plotting
plt.figure(figsize=(10, 6))
plt.plot(x_new2, y2, label='Function')
plt.axhline(xmin=-4, xmax=4, y=0, c='r', label='x-axis')
plt.scatter(val_1, f2(val_1), marker='o', color='blue', s=100, label='Solution')
plt.scatter(val_2, f2(val_2), marker='o', color='blue', s=100)
plt.xlabel('x values')
plt.ylabel('y values')
plt.legend()
plt.show()
```

### Newton Method Example

```python
## Define the Function and plot it  --> () = ^+.
def f3(x):
    return x**3 - 2*x + 2

x_new3 = np.linspace(-2, 2)
y3 = f3(x_new3)

## plotting
plt.figure(figsize=(10, 6))
plt.plot(x_new3, y3, c='g', label='Function')
plt.axvline(ymin=-2, ymax=6, x=0, c='r', label='y axis')
plt.axhline(xmin=-2, xmax=2, y=0, c='b', label='x axis')
plt.scatter(-2, f3(-2), c='k')
plt.scatter(-1.8, f3(-1.8), c='k')
plt.scatter(-1.77, f3(-1.77), c='k', label='steps')
plt.xlabel('x values')
plt.ylabel('y values')
plt.legend()
plt.show()
```
